```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib
```

```python
matplotlib.rcParams['font.size'] = 8
matplotlib.rcParams['figure.figsize'] = (12, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
DF_PV = pd.read_csv('/content/sample_data/CleanDataforVisualisation.csv')
DF_PV
```

```python
DF_PV['Financial Year'].unique()
```

```python
DF_PV['Offence Group'].unique()
```

```python
sns.set(style="whitegrid")
sns.scatterplot(x=DF_PV['Financial Year'], y=DF_PV['Number of Offences'], hue=DF_PV['Offence Group'], s=100)
```

```python
Offence_Group = DF_PV.groupby("Offence Group")["Number of Offences"].sum()
Offence_Group
```

```python
Offence_Group.plot(kind='bar', ylabel='Total Number of Offences')
```

```python
Years_Offence = DF_PV.groupby("Financial Year")["Number of Offences"].sum()
```

```python
plt.plot(Years_Offence)
plt.xlabel('Financial Year')
plt.ylabel('Total Offences')
```

```python
sns.barplot(x='Financial Year', y='Number of Offences', hue = 'Financial Quarter', data=DF_PV);
```

```python
#Plot Multiple Together
fig, axes = plt.subplots(2, 2, figsize=(20, 12))

sns.set(style="whitegrid")
sns.scatterplot(x=DF_PV['Financial Year'], y=DF_PV['Number of Offences'], hue=DF_PV['Offence Group'], s=100, ax=axes[0,0])


Offence_Group.plot(kind='bar', ylabel='Total Number of Offences', ax=axes[0,1])

plt.plot(Years_Offence)
plt.xlabel('Financial Year')
plt.ylabel('Total Offences')
ax=axes[1,0]

sns.barplot(x='Financial Year', y='Number of Offences', hue = 'Financial Quarter', data=DF_PV, ax=axes[1,0])

plt.tight_layout(pad=2)
```
