```python
import pandas as pd
```

```python
DF_train = pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
DF_train
```

```python
DF_test = pd.read_csv('/content/sample_data/california_housing_test.csv')
```

```python
DF_test
```

```python
x_test = x_train = DF_test.iloc[:, 2:8]
```

```python
y_test = DF_test.iloc[:, 8:9]
```

```python
DF_train_final = DF_train.iloc[:, 0:]
```

```python
DF_train_final
```

```python
x_train = DF_train_final.iloc[:, 2:8]
```

```python
x_train
```

```python
y_train = DF_train_final.iloc[:, 8:9]
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
model_LR = LinearRegression()
model_LR.fit(x_train, y_train)
```

```python
y_predict = model_LR.predict(x_test)
```

```python
y_predict
```

```python
from sklearn import metrics
import numpy as np
print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_predict))
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_predict))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
```

```python
#Support Vector Regression
```

```python
from sklearn.svm import SVR
model_SVR = SVR(kernel='linear')
model_SVR.fit(x_train, y_train)
```

```python
y_predict2 = model_SVR.predict(x_test)
```

```python
y_predict2
```

```python
print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_predict2))
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_predict2))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_predict2)))
```

```python
#RBF
```

```python
from sklearn.svm import SVR
model_SVR_RBF = SVR(kernel = 'rbf')
model_SVR_RBF.fit(x_train, y_train)
```

```python
y_predict3 = model_SVR_RBF.predict(x_test)
y_predict3
```

```python
print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_predict3))
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_predict3))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_predict3)))
```

```python
#Poly
```

```python
from sklearn.svm import SVR
model_SVR_poly = SVR(kernel = 'poly')
model_SVR_poly.fit(x_train, y_train)
```

```python
y_predict4 = model_SVR_poly.predict(x_test)
y_predict4
```

```python
print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_predict4))
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_predict4))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_predict4)))
```
